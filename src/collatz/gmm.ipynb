```python
from sklearn.mixture import GaussianMixture

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
sns.set()
```

```python
xs = np.array([int(l.strip()) for l in open('out.txt').readlines()]).reshape(-1, 1)
plt.hist(xs[:, 0], bins=32, histtype='stepfilled')
pass
len(xs)
```

```python
gm = GaussianMixture(n_components=2, max_iter=1000)
gm.fit(xs)

x = np.linspace(-100, 400, 1000)

logprob = gm.score_samples(x.reshape(-1, 1))
responsibilities = gm.predict_proba(x.reshape(-1, 1))
pdf = np.exp(logprob)
pdf_individual = responsibilities * pdf[:, np.newaxis]

plt.hist(xs, 32, density=True, histtype='stepfilled', alpha=0.4)
plt.plot(x, pdf, '-k', c='blue', linewidth=2)
plt.plot(x, pdf_individual, '--k', c='k', alpha=0.4)

pass
```
